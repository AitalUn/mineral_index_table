```python
import pandas as pd

df = pd.read_excel("indicies.xlsx")

data = []
for abbr, formula in zip(df['abbreviation'], df['LANDSAT-89 OLITIRS']):
    if type(formula)==float or formula == "constant":
        continue
    data.append({"name": abbr, "formula": formula})
```

```python
a = pd.DataFrame(data)
a.to_excel('landsat_index.xlsx')
```
