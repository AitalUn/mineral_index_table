```python
import pandas as pd
from pathlib import Path
import sys
import os

path = Path(os.getcwd()).parent / 'final_delivery' / 'indicies.xlsx'
```

```python
df = pd.read_excel(path)
data = []
for name, formula, d, l, r in zip (df['abbreviation'], df['Modis'], df['description'], df['idb_link'], df['russian']):
    if type(formula) is float:
        continue
    data.append({
        "link": l,
        "name": name, 
        "formula": formula, 
        "description": d,
        "ru name": r})
```

```python
pd.DataFrame(data).to_excel("modis_index_list.xlsx")
```
