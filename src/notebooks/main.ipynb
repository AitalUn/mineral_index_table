```python
from mathml_to_latex.converter import MathMLToLaTeX


import sys
import os
from pathlib import Path
import pandas as pd

import requests

from tqdm import tqdm
from bs4 import BeautifulSoup

from mathml_to_latex.converter import MathMLToLaTeX


root = Path(os.getcwd()).parent.parent

excel_path = root / r"data\excel\IndexApplications.xlsx"
root_link = "https://www.indexdatabase.de"


def parse_index_page(link_of_index_details_page: "str"):
    response = requests.get(link_of_index_details_page)
    response.raise_for_status()
    page_of_index = BeautifulSoup(response.text, "html.parser")

    index_table_soup = page_of_index.select("div.table.vektor tr")
    try:
        reference_table_soup = page_of_index.select("div.table.matrix")[2]
    except:
        reference_table_soup = None

    ref_data = []
    if reference_table_soup:
        article_titles = [r.get_text(strip=True) for r in reference_table_soup.select("td")[1:]]
        article_links = [f"{root_link}\{a['href']}" for td in reference_table_soup.select("td") 
         for a in td.find_all('a') 
         if a.get('href')]

        for i, link in enumerate(article_links):
            if i > 4:
                break
            
            response = requests.get(link)
            response.raise_for_status()
            artcle_page_soup = BeautifulSoup(response.text, 'html.parser')
            trs = artcle_page_soup.select("div.table.vektor table.vektor tr")
            article_data = {
                "author": trs[0].get_text(strip=True),
                "title": trs[1].get_text(strip=True),
                "journal": trs[2].get_text(strip=True),
                "doi": trs[3].get_text(strip=True),
                "link": trs[4].get_text(strip=True)[4:],
                "date": trs[5].get_text(strip=True)[3:],
            }

            ref_data.append(article_data)
    else:
        print("для данного индекса нету статьи")


    index_data = {
        "name": index_table_soup[0].select("td")[0].get_text(strip=True),
        "abbreviation": index_table_soup[1].get_text(strip=True)[12:],
        "formula": MathMLToLaTeX().convert(str(index_table_soup[2].select("td")[0].find("math"))),
        "variables": index_table_soup[3].get_text(strip=True)[9:],
        "explanation_of_variables": index_table_soup[4].get_text(strip=True)[18:],
        "description": index_table_soup[7].get_text(strip=True)[11:],
        "wavelengths": index_table_soup[5].get_text(strip=True),
        "source": index_table_soup[6].get_text(strip=True)[4:],
        "article_name": "\n".join([f"{i+1}.{r['author'][6:]} - {r['title']}" for i, r in enumerate(ref_data)]),
        "source": "\n".join([f"{i+1}.{r['link']}" for i, r in enumerate(ref_data)]),
        "doi": "\n".join([f"{i+1}.{r['doi'][3:]}" for i, r in enumerate(ref_data)])
    }

    # print(index_data)

    return index_data


```

```python

# Читаем файл
df = pd.read_excel(excel_path)
df['Link']


index_list = []
for link_of_application in tqdm(df['Link']):

    response = requests.get(link_of_application)
    response.raise_for_status()
    page_html = response.text
    table_of_application_soup = BeautifulSoup(page_html, 'html.parser').select("div.table.matrix table.matrix")[0].select("tr")[1:]
    

    for index_soup in tqdm(table_of_application_soup, desc="Индексы", leave=True, position=1):
        link_of_index = root_link + index_soup.select("td")[1].select("a")[0]['href']

        try:
            index_list.append(parse_index_page(link_of_index_details_page=link_of_index))
        except:
            pass


```

```python
df = pd.DataFrame(index_list)
df.to_excel('index_db.xlsx', index=False, engine="openpyxl")
```
