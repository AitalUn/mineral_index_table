```python
from pathlib import Path
import os
import pandas as pd
import ast


existing_articles = []
papers_dir = Path(os.getcwd()).parent.parent / "data" / "papers"
for i, a in enumerate(papers_dir.iterdir()):
    existing_articles.append(a.name)

df = pd.read_excel("index_table_v4.ods")

required_articles = set()
for a in df['references']:
    try:
        adict = ast.literal_eval(a)
        for article in adict:
            required_articles.add(article['article_name'])
    except:
        print(a)
```

```python
from difflib import SequenceMatcher, get_close_matches

dont_exist = []
treshold = 0.9
for a in required_articles:
    max_ratio = 0
    for e in existing_articles:
        ratio =SequenceMatcher(None, a, e).ratio()  
        if ratio > max_ratio:
            max_ratio = ratio
    if max_ratio > treshold:
        print(f"{a}{e}\n")
    else:
        dont_exist.append(a)

```
