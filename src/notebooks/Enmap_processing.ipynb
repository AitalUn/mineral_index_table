```python
from pathlib import Path
import pandas as pd
import os

sheets = ["VNIR", "SWIR"]
bands = {}

enmap_dir = Path(os.getcwd()).parent.parent /"data"/"excel"/"EnMAP_Spectral_Bands_update.xlsx"
for s in sheets:
    data = pd.read_excel(enmap_dir, sheet_name=s)
    
    for i, cw, fw in zip(data["BAND #"] ,data["CW (nm)"], data["FWHM (nm)"]):
        bands[f"{s}_{i}"] = (round(cw-fw/2, ndigits=2),round(cw+fw/2, ndigits=2))

```

```python
import json

with open("EnMAP.json", "w", encoding="utf8") as f:
    json.dump(bands, f, indent=4)
```
