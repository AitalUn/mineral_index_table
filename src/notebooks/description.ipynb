```python
from pathlib import Path
import os

path = Path(os.getcwd()).parent.parent/"data"/"excel"/"abstract_text.txt"
with open(path, "r", encoding="utf-8") as f:
    text = f.read()

chunks = text.split("#")
```

```python
import requests
from pydantic import BaseModel

class ResponseModel(BaseModel):
    name: str
    description: str



def generate_description(prompt: str):
    response = requests.post(
        url="https://api.openai.com/v1/chat/completions",
        headers={
            "Authorization": f"Bearer {key}",
            "Content-Type": "application/json"
        },
        json={
            "model": "gpt-5-nano",  # Или gpt-4o-mini
            "messages": [
                {
                    "role": "system", 
                    "content": "Ты пишешь описания для аббревиатур минеральных инедксов на основе предоставленного контекста, в ответе опиши для чего применяются индексы. Если в контексте не хватает информации, ответь 'Не хватает данных'."
                },
                {
                    "role": "user", 
                    "content": prompt
                }
            ],
            "temperature": 1,

        }
    )
    return response.json()['choices'][0]['message']['content']
```

```python
from tqdm import tqdm

for i, abrev in tqdm(enumerate(df['abbreviation'])):
    context = ""
    if abrev is None:
        continue

    for chunk in chunks:
        if str(abrev) in chunk:
            context += chunk

    if context == "":
        continue
    response = generate_description(prompt=f"Сгенерируй краткое описание для аббревиатуры {abrev} на основе следующего контекста: {context}")
    print(f"{i}. {response}")
    df.at[i, 'generated_description'] = response
    # df.at[i, 'russian_name'] = response["name"]
```
