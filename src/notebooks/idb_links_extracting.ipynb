```python
import sys
import os
import requests
import pandas as pd

from pathlib import Path
from tqdm import tqdm
from bs4 import BeautifulSoup


root = Path(os.getcwd()).parent.parent

excel_path = root / r"data/excel/IndexApplications.xlsx"
root_link = "https://www.indexdatabase.de"

df = pd.read_excel(excel_path)
```

```python
target_links = set()

for link in df['Link']:
    response = requests.get(link)
    response.raise_for_status()
    html = response.text
    soup = BeautifulSoup(html)
    
    table = (soup.select("div#content table.matrix")[0].select("tr")[1:])
    links = [f"{root_link}/{r.select("td")[1].select("a")[0]['href']}" for r in table]
    for link in links:
        target_links.add(link)
```

```python
index_list = pd.DataFrame(target_links)
```

```python
index_list.to_excel('index_links.xlsx', index=False)
```
